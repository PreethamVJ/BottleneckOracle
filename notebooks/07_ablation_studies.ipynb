## 7. Summary & Gate Check

Evaluate if GNN results meet checkpoint criteria

```python
def checkpoint_gate_evaluation(ablation_results):
    """
    W5 CHECKPOINT: Evaluate if GNN results are solid enough to proceed to W6 LLM work.
    
    GATE: if GNN is shaky, skip W6 LLM and use W6–W7 for more ablations instead.
    """
    gate_criteria = {
        'directed_spearman_ok': ablation_results['ablation_a']['directed']['spearman_rho'] > 0.7,
        'delta_t_error_ok': ablation_results['ablation_a']['directed']['delta_T_error'] < 0.05,
        'what_if_acceptable': len([r for r in ablation_results['what_if'] if r['error_percent'] < 10]) >= 8,
        'ood_degradation_acceptable': ablation_results['ood']['mae_degradation_percent'] < 20
    }
    
    proceed_to_llm = all(gate_criteria.values())
    
    print("W5 Checkpoint Gate Results:")
    for criterion, passed in gate_criteria.items():
        status = "PASS" if passed else "FAIL"
        print(f"  {criterion}: {status}")
    
    if proceed_to_llm:
        print("\n✓ GATE PASSED: Proceed to Sprint 6 (GraphRAG/LLM)")
    else:
        print("\n✗ GATE FAILED: Use W6-W7 for more ablations instead of LLM work")
    
    return proceed_to_llm, gate_criteria

# Evaluate checkpoint gate
proceed_to_llm, gate_criteria = checkpoint_gate_evaluation(ablation_results)
```
